# Data Dictionary

The following variables are included in the dataset:

* Campaign_ID – Unique identifier assigned to each marketing campaign.
* Campaign_Type – The type of marketing strategy used for the campaign (e.g., Influencer, Social Media, Paid Ads, Email, SEO).
* Target_Audience – The specific audience group targeted by the marketing campaign.
* Duration – The length of the campaign in days.
* Channel_Used – The marketing platforms used to run the campaign (e.g., Instagram, Google, YouTube, Email, WhatsApp, Facebook).
* Impressions – The total number of times the advertisement was shown to users.
* Clicks – The number of times users clicked on the advertisement.
* Leads – The number of users who showed interest or performed a lead-generating action after clicking the ad.
* Conversions – The number of users who completed the desired action (such as making a purchase).
* Revenue – The total revenue generated from the campaign.
* Acquisition_Cost – The cost incurred to acquire customers through the campaign.
* ROI (Return on Investment) – The profitability of the campaign calculated based on revenue generated relative to cost.
* Language – The language used in the campaign content.
* Engagement_Score – A score representing how actively users interacted with the campaign content.
* Customer_Segment – The category of customers targeted by the campaign (e.g., Working Women, Premium Shoppers, Youth).
* Date – The date on which the campaign performance data was recorded.

**Check funnel by channel.**

```python
df.groupby('Channel_Used')[['Impressions','Clicks','Leads','Conversions']].sum().head()
```

**We need conversion rates, not just totals**

```python
channel_funnel = df.groupby('Channel_Used')[['Impressions','Clicks','Leads','Conversions']].sum()

channel_funnel['CTR'] = channel_funnel['Clicks'] / channel_funnel['Impressions']
channel_funnel['Lead_Rate'] = channel_funnel['Leads'] / channel_funnel['Clicks']
channel_funnel['Conversion_Rate'] = channel_funnel['Conversions'] / channel_funnel['Leads']

channel_funnel.sort_values('Conversion_Rate', ascending=False).head(10)
```

Top channel combinations show very similar funnel performance:

CTR: ~8–9%

Lead Rate: ~38–41%

Conversion Rate: ~57–59%

So the funnel is fairly consistent across channels.

**Analyze funnel by Campaign_Type**

```python
impressions = df['Impressions'].sum()
clicks = df['Clicks'].sum()
leads = df['Leads'].sum()
conversions = df['Conversions'].sum()

drop_click = impressions - clicks
drop_lead = clicks - leads
drop_conversion = leads - conversions

drop_click, drop_lead, drop_conversion
```

**Clicks → Leads**

* Drop: 156,153,960

**Leads → Conversions**

Many people see the campaign but don’t click.

This suggests:

**ROI vs Acquisition Cost**

High ROI mostly happens at low acquisition cost

Key marketing insight
* Spending more money on acquiring customers does not guarantee better returns.

Efficient campaigns are those that:
* keep acquisition cost low
* still generate good conversions and revenue

**Business recommendation**

Companies should:
* optimize targeting and creatives
* avoid overspending on expensive campaigns
* scale campaigns that show low cost + high ROI

# Conclusion & Recommendations

* Improve Top-of-Funnel Engagement
The largest drop-off occurs between impressions and clicks (~8.5% CTR). Improving ad creatives, messaging, and audience targeting could increase engagement.

* Prioritize High-ROI Campaigns
Although Influencer campaigns generate the most revenue, Social Media campaigns deliver the highest ROI, making them more cost-efficient for marketing investment.

* Focus on High-Value Customer Segments
Campaigns targeting Working Women generate the highest revenue, while Premium Shoppers show the highest conversion rates. These segments should be prioritized for targeted marketing.

* Leverage Multi-Channel Marketing Strategies
Campaigns using multiple channels (e.g., Email + Google + YouTube) show higher conversion rates and ROI compared to single-channel campaigns.

* Optimize Acquisition Costs
Higher acquisition costs often lead to lower ROI. Marketers should identify and scale low-cost, high-performing campaigns.

* Increase Engagement to Boost Conversions
Higher engagement scores are associated with increased conversions, indicating that improving user interaction with marketing content can enhance funnel performance.
